```
from os import listdir
from PIL import Image
import random
import numpy as np
from zipfile import ZipFile 
import datetime
from timeit import default_timer as timer
import cv2
import os
```

```
bgpath = '/content/gdrive/My Drive/Session15/bg224/'
fgpath = '/content/gdrive/My Drive/Session15/Foreground/'
bglist = listdir(bgpath)
count = 1
# for bgimage in bglist:
out_zip =  ZipFile('bg_fg_bg_mask.zip', mode='a')
# ZipFile('depth_fg_bg.zip', mode='a', compression=zipfile.ZIP_STORED)        
start = timer()
for ind in range(1,101):
    now = datetime.datetime.now()
    print('state time',now)
    bg = Image.open(bgpath +str(ind)+".jpeg")
    # bg = bgpathall[ind]
    fglist = listdir(fgpath)
    print(ind)
    for fgimage in fglist:
        fg = Image.open(fgpath + fgimage)
        # fg = fgpathall[fgind]
        w = int((fg.width/fg.height)*160)
        fg = fg.resize((w,160))
        
        for x in range(2):
            fg_bg = Image.new('RGBA', (bg.width,bg.height), (0, 0, 0, 0))
            mask = Image.new('RGB', (bg.width,bg.height), (0, 0, 0, 1))
            fg_bg.paste(bg, (0,0))
            xy = (random.randint(1, 224-w),random.randint(1, 112)) 
            fg_bg.paste(fg, xy, mask=fg)
            mask.paste(fg,xy , mask=fg)
            # fg_bg.save("fg_bg/fg_bg"+str(count)+".png", format="png")
            
            ar1 = np.array(mask)
            ar1[ar1 > 0] = 255
            mask = Image.fromarray(ar1)
            # mask.save("Mask_fg_bg/mask_fg_bg"+str(count)+".png", format="png")
            fg_bg = fg_bg.convert("RGB");
            fg_bg.save('fg_bg.jpeg', format="jpeg", optimize=True, quality=35)
            mask.save('mask.jpeg',  format="jpeg",optimize=True, quality=95)
            # save the files to .zip file
            out_zip.write(bgpath +str(ind)+".jpeg", f'bg/'+str(count)+'.jpeg')
            out_zip.write('fg_bg.jpeg', f'fg_bg/'+str(count)+'.jpeg')
            out_zip.write('mask.jpeg', f'mask_fg_bg/'+str(count)+'.jpeg')
            count += 1
    print(datetime.datetime.now()-now)
out_zip.close()
duration = timer() - start
print(duration)                
```

```
def load_images_from_folder(folder):
    images = []
    for filename in listdir(folder):
        img = Image.open(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```
